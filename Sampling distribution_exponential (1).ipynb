### Import required libraries

```python
import random
import numpy as np
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import math
```

### Function to create bar chart

```python
def create_bar_chart(title,x_values, y_values,x_label='Value', y_label='Frequency'):
       """
    Display a bar chart
    @param title the chart title.
    @param x_label the label for the x axis
    @param y_label the label for the y axis
    @param x_values the x values to plot
    @param y_values the y values to plot
    """
    axes = sns.barplot(x_values, y_values,color='b')
    axes.set_title(title)
    axes.set(xlabel=x_label)
    axes.set(ylabel=y_label)
    sns.set(rc={'figure.figsize':(25,8)})
    sns.set_style('whitegrid')
    sns.set_context('poster')
```

### Create uniform population distribution by generating uniform random values
A normal population distribution is used to undertake this sampling distribution experiment. The distribution consists of normally distributed random values.

```python
def population_distribution(mu,sigma,pop_size):
    nvs=[int(random.normalvariate(mu,sigma))
                              for _ in range(pop_size)]
    n_value,n_frequency=np.unique(nvs,return_counts=True)
    create_bar_chart(f'Population of {pop_size} integers with mean={statistics.mean(nvs):.2f}, median={statistics.median(nvs)}, std dev={statistics.pstdev(nvs):.2f}',n_value,n_frequency)
    plt.show()
    return(nvs)
    
```

A population size of 2000000 and mean 12 and standard deviation 4.1 is chosen for the experiment

```python
nvs=population_distribution(12.00,4.1,2000000)
```

### Sampling distribution of means
Here we generate a sample distribution of means by taking random samples from the population, calculating their means and generating a distribution of means.

```python
def sample_mean(population,sample_size, sample_count):
        
    """ Generate sampling distribution of means 
    
    
    @param population is population distribution 
    @param sample_size is the number of samples 
    @param sample_count number of times samples are chosen

    
    """
    mean_sample=[int(statistics.mean(random.sample(population,sample_size))) for _ in range(sample_count)]
    sdm_value,sdm_freq=np.unique(mean_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Means, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(mean_sample):.2f}, standard error={std_err:.2f}',sdm_value,sdm_freq)
    plt.show()
```

#### Graphs of different sample count
In this experiment, we first vary the sample count to study the variation in sampling distribution due to changes in sample count. We chose three different sample counts: 100, 1000 and 10000

```python
sample_mean(nvs,30,10000)
```

```python
sample_mean(nvs,100,10000)
```

```python
sample_mean(nvs,10,10000)
```

#### Inference
From the above graphs, we can see that as the sample size increases, the sampling distribution starts resembling normal distribution more and more. We can see that sampling distribution with sample size 100 has mean value 11.01 closest to population mean 11.50 and syandard error has decreased.



```python
sample_mean(nvs,30,100)
```

```python
sample_mean(nvs,30,1000)
```

```python
def sample_median(population,sample_size, sample_count):
    median_sample=[int(statistics.median(random.sample(population,sample_size))) for _ in range(sample_count)]
    sdmd_value,sdmd_freq=np.unique(median_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Medians, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(median_sample):.2f}, standard error={std_err:.2f}',sdmd_value,sdmd_freq)
    plt.show()
```

```python
sample_median(nvs,30,10000)
```

#### Inference
From the above graphs, we can see that as the sample count increases, the sampling distribution starts resembling normal distribution more and more. We can see that sampling distribution with sample size 10000 has mean value 11.35 closest to population mean 11.50 and standard error has decreased.


```python
sample_median(nvs,10,10000)
```

```python
sample_median(nvs,100,10000)
```

```python
sample_median(nvs,30,100)
```

```python
sample_median(nvs,30,1000)
```

#### Inference
From the above graphs, we can see that as the sample count and sample size increase, the sampling distribution starts resembling normal distribution more and more. We can see that sampling distribution with sample count 1000 and sample size 30 has mean value 11.35 closest to population mean 11.50 and standard error has decreased.


```python
def sample_stddev(population,sample_size, sample_count):
    stddev_sample=[int(statistics.pstdev(random.sample(population,sample_size))) for _ in range(sample_count)]
    stdev_value,stdev_freq=np.unique(stddev_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Standard Deviations, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(stddev_sample):.2f}, standard error={std_err:.2f}',stdev_value,stdev_freq)
    plt.show()
```

```python
sample_stddev(nvs,10,10000)
```

#### Graphs of different sample counts:
Here, we vary the sample count to study the variation in sampling distribution of std deviations. We chose sample size 10, 30, 100 and kept the sample size constant.

```python
sample_stddev(nvs,30,10000)
```

```python
sample_stddev(nvs,100,10000)
```

#### Inference:
From the above graphs, we see that sampling distributions of std deviations get centered around the population std deviation value as the sample size increase. The sample size of 100 sampling distribution appears closest to normal distriution and also has sample peak closest to population std deviation.




```python
sample_stddev(nvs,30,100)
```

#### Inference:
We can infer from the above graphs that as we increase sample count, the distribution approaches normal distribution. The sample count 20,000 has closest resmblence to normal distribution.
