```python
%matplotlib inline
```

```python

fig=vehicle_certification_data.groupby(['Manufacturer', 'Fuel Type']).size().unstack().plot(kind='bar', stacked=True)

plt.ylabel('Count')
plt.title('Count of UK car models by manufacturers by fuel type')
plt.rcParams["figure.figsize"] = (20,10)
```

```python
vehicle_certification_data.groupby('Fuel Type')['Engine Capacity'].mean().dropna().plot(kind='bar')
plt.ylabel('Mean Engine Capacity (cc)')
plt.title('Engine capacity by fuel type')
```

```python
vehicle_certification_data.groupby('Fuel Type')['Emissions NOx [mg/km]'].mean().plot(kind='bar')
plt.ylabel('Mean NOx emissions (mg/km)')
plt.title('NOx Emissions by fuel type')
```

```python
vehicle_certification_data.groupby('Manufacturer')['Emissions CO [mg/km]'].mean().plot(kind='bar')
plt.ylabel('Average CO Emissions (mg/km)')
plt.title('CO Emissions by Manufacturers')
plt.rcParams["figure.figsize"] = (20,10)
```

```python
vehicle_certification_data.groupby('Manufacturer')['Electric energy consumption Miles/kWh'].mean().dropna().plot(kind='bar')
plt.ylabel('Average electricity consumption(miles/kWh)')
plt.title('Electric car energy consumption by manufacturer')
```
