### Visualize the data

#### Displaying the images of 40 different subjects

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# Arrange the images in a 4 by 6 grid.
figure, axes = plt.subplots(nrows=4, ncols=10, figsize=(10, 6))

# Display each image and its target, without the axes labels.
for item in zip(axes.ravel(), faces.images[::10], np.unique(faces.target)):
    axes, image, target = item
    axes.imshow(image,cmap=plt.cm.bone)
    axes.set_xticks([])  # remove x-axis tick marks
    axes.set_yticks([])  # remove y-axis tick marks
    axes.set_title(target)
    
plt.tight_layout()     
```

### Split the data for training and testing

##### In order to cover all the subjects in our testing set, we set test size as 40%

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     faces.data, faces.target, test_size=0.40, random_state=11)

print(f'X_train.shape = {X_train.shape}')
print(f'X_test.shape  = {X_test.shape}')

print()
print(f'y_train.shape = {y_train.shape}')
print(f'y_test.shape  = {y_test.shape}')
```

### Create and train the k-nearest neighbors model

##### Using the training data to create and train a k-nearest neighbor model. We set k value as 5

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X=X_train, y=y_train)
```

### Test the model by predicting digit classes 

##### Using the testing data to make some predictions

```python
predicted = knn.predict(X=X_test)
expected = y_test

print('First twenty predictions:')
print(f'predicted[:20] = {predicted[:20]}')
print(f' expected[:20] = {expected[:20]}')
```

```python
wrong = [ (pred, exp) 
          for (pred, exp) in zip(predicted, expected) 
          if pred != exp
        ]

print('Wrong predictions:')
print(wrong)
print()
print(f' The number of wrong predictions :{len(wrong)}')
```

```python
print('Wrong predictions:')
print()


for data, pred, exp in zip(X_test, predicted, expected):
    if pred != exp:
        figure = plt.figure(figsize=(1, 1))
        image = data.reshape(64,64)
        plt.imshow(image,cmap=plt.cm.bone)
        plt.xticks([])  # remove x-axis tick marks
        plt.yticks([])  # remove y-axis tick marks
        plt.title(f'pred = {pred}, exp = {exp}')   
```

### Model prediction accuracy

```python
print(f'Prediction accuracy score = {knn.score(X_test, y_test):.2%}')
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_true=expected, y_pred=predicted)

print('Confusion matrix:')
print(confusion)
```

```python
import pandas as pd
import seaborn as sns
plt.figure(figsize=(16, 9))
confusion_df = pd.DataFrame(confusion, index=range(40), columns=range(40))
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

### Classification report

```python
from sklearn.metrics import classification_report

names = [str(face) for face in faces.target[::10]]
print(classification_report(expected, predicted,target_names=names))
```
