# Seaborn Charts of Normally Distributed Random Values

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(20,8)})
```

```python
import random 
```

## Generating 1000 normally distributed random values over the range 0-25

```python
normal_values=[int(random.normalvariate(12.5,4.1)) for _ in range(1000)]
```

```python
values_normal,frequencies_normal=np.unique(normal_values,return_counts=True)
```

### Barchart

```python
axes=sns.barplot(x=values_normal, y=frequencies_normal,color='#95a5a6')
axes.set_title('Normal Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
axes.set(xlim=(-1,25))
sns.set_style('whitegrid')
sns.set_context('poster')
```

### Univariate distribution plot

```python

axes=sns.distplot(normal_values,color='#95a5a6')
axes.set_title('Normal Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
axes.set(xlim=(0,25))
```
