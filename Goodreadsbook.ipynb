```python
good_read_books.groupby('authors')['average_rating'].describe().fillna(0).round(3) #displaying the authors statistics by average ratings #filling NaN values with 0
```

```python
good_read_books.groupby('language_code')['average_rating'].describe().fillna(0).round(3) #displaying average ratings statistics by language codes
```

```python
good_read_books.groupby('language_code')['ratings_count'].describe().fillna(0).round(3) #displaying rating count statistics by language codes
```

```python
good_read_books.groupby('language_code')['text_reviews_count'].describe().fillna(0).round(3) #displaying reviews count statistics by language codes
```

```python
good_read_books.groupby('language_code')['# num_pages'].describe().fillna(0).round(3) #displaying number pages statistics by language codes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
good_read_books.groupby('language_code')['average_rating'].mean().plot(kind='bar') #plotting average ratings of books in differnt languages by language codes
plt.xlabel('Language Code')
plt.ylabel('Average Rating')
plt.title('Average rating of books in different languages')
```

```python
good_read_books.groupby('language_code')['text_reviews_count'].mean().plot(kind='bar') #plotting average number of text reviews for books in differnt languages by language codes
plt.xlabel('Language Code')
plt.ylabel('Average text reviews')
plt.title('Average number of text reviews for books in different languages')
```

```python
good_read_books.groupby('language_code')['ratings_count'].mean().plot(kind='bar') #plotting average number of ratings count in differnt languages by language codes
plt.xlabel('Language Code')
plt.ylabel('Average ratings count')
plt.title('Average ratings count for books in different languages')
```

```python
good_read_books.groupby('language_code')['bookID'].count().plot(kind='bar') #plotting average number of text reviews in differnt languages by language codes
plt.xlabel('Language Code')
plt.ylabel('Average text reviews')
plt.title('Average number of text reviews for books in different languages')
```

```python
good_read_books.groupby('language_code')[['authors']].count() #displaying number of Authors in differnt languages of books using language code
```

```python
good_read_books.groupby('language_code')[['authors']].count().plot(kind='bar') # plotting the average number of authors in different languages of books by language code
plt.xlabel('Language Code')
plt.ylabel('Authors')
plt.title(' number of authors in different languages')
```
