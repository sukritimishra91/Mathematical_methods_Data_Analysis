
Seaborn Charts of Uniform Random Values

```python
import random
import seaborn as sns #import the necessary libraries
import numpy as np

```

Generating 1000 Uniform Random Values between 0 and 25



```python
normal_values=np.random.normal(loc=12.5, scale=6.25, size=1000)
mean_uni=np.mean(normal_values).round(1)
mean_uni
```

Univariate distribution plot (with histogram and kernal density estimates)


```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)# size of the graph
```

```python
from scipy.stats import norm #statistic function for normal distribution graph
normal_values = np.random.normal(size=10000) 
sns.distplot(normal_values, kde=True, bins=10); #Histograms and KDE can be combined using distplot;kde= kernel density estimation 
plt.xlabel('Random Variables')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
```

Barchart


```python
sns.distplot(normal_values, kde=False, bins= 20); #bins:divide the entire range of values into a series of intervals. 
plt.xlabel('values')
plt.ylabel('Frequency')
plt.title("Normal Distribution Histogram ")
```

Line Graph

```python
import numpy
import matplotlib.pyplot as plt
uniform_values=[int(random.normal(0,26)) for _ in range(1000)]
uniform_values = numpy.random.uniform(low=0,high=25,size=100)
plt.plot(uniform_values)
plt.xlabel('size')
plt.ylabel('range')
plt.title("Normal Distribution Histogram ")
plt.show() # Show the plot


```
