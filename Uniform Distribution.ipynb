```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
sns.set_style('whitegrid')
values = (np.random.uniform(size=1000, low=0, high = 25))
sns.distplot(values, kde=True, bins=10);
plt.xlabel('Random Variables')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')

```

```python
import numpy
import matplotlib.pyplot as plt
values = numpy.random.uniform(low=0,high=25,size=100)
plt.plot(values)
plt.show()
```

```python
plt.hist(values, histtype = 'bar', rwidth= 0.5)
plt.xlabel('variable')
plt.ylabel('frequency')
plt.title('Histogram Graph')
        (xlim=(0,25))
sns.set_context('notebook', font_scale=1.0, rc={"lines.linewidth": 1.5})
plt.show()
```
