### Import required libraries

```python
import random
import numpy as np
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import math
```

### Function to create bar chart

```python
def create_bar_chart(title,x_values, y_values,x_label='Value', y_label='Frequency'):
    """
    Display a bar chart
    @param title the chart title.
    @param x_label the label for the x axis
    @param y_label the label for the y axis
    @param x_values the x values to plot
    @param y_values the y values to plot
    """
    axes = sns.barplot(x_values, y_values,color='b')
    axes.set_title(title)
    axes.set(xlabel=x_label)
    axes.set(ylabel=y_label)
    sns.set(rc={'figure.figsize':(25,8)})
    sns.set_style('whitegrid')
    sns.set_context('poster')
```

### Create uniform population distribution by generating uniform random values

A uniform  population distribution is used to undertake the sampling distribution experiment. The distribution consists of random values of range 0-39.



```python
def population_distribution(pop_size):
    uvs = [int(random.uniform(0, 40)) 
                              for _ in range(pop_size)]
    u_value,u_frequency=np.unique(uvs,return_counts=True)
    create_bar_chart(f'Population of {pop_size} integers with mean={statistics.mean(uvs):.2f}, median={statistics.median(uvs)}, std dev={statistics.pstdev(uvs):.2f}',u_value,u_frequency)
    plt.show()
    return(uvs)
```

The population size chosen for the distribution is 2000000

### Display the population distribution bar plot
The population mean is 19.5, median is 20 and standard deviation is 11.54

```python
uvs=population_distribution(2000000)
```

### Generate sampling distribution of means


```python
def sample_mean(population,sample_size, sample_count):
    
    """ Generate sampling distribution of means 
    
    
    @param population is population distribution 
    @param sample_size is the number of samples 
    @param sample_count number of times samples are chosen

    
    """
    mean_sample=[int(statistics.mean(random.sample(population,sample_size))) for _ in range(sample_count)]
    sdm_value,sdm_freq=np.unique(mean_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Means, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(mean_sample):.2f}, standard error={std_err:.2f}',sdm_value,sdm_freq)
    plt.show()

    
```

### Sampling distribution of means
Here we generate a sample distribution of means by taking random samples from the population, calculating their means and generating a distribution of means.

#### Graphs of different sample count
In this experiment, we first vary the sample count to study the variation in sampling distribution due to changes in sample count. We chose three different sample counts: 100,1000 and 10000


```python
sample_mean(uvs,30,100)
```

```python
sample_mean(uvs,30,1000)
```

```python
sample_mean(uvs,30,10000)
```

#### Inference: 
From the above graphs, we can see that as the sample count increases, the sampling distribution starts resembling normal distribution more and more. We can see that sampling distribution with sample count 1000 has mean value (19.12) closest to population mean (19.5).

#### Graphs of different sample sizes 
In this experiment, we vary the sample size to see its effect on sampling distribution of means. We chose sample sizes 10, 100 and 200 to study the effect of sample size.

```python
sample_mean(uvs,10,10000)
```

```python
sample_mean(uvs,100,10000)
```

```python
sample_mean(uvs,200,10000)
```

#### Inference:
From the above graphs, we can infer that as sample size increases, the values get more and more centered around the population mean. The standard error decreases with increasing sample size. One interesting inference is that sampling distribution is closest to population mean for sample size 10, although the satndard error is highest.


### Sampling distribution of medians
Here we generate a sample distribution of medians by taking random samples from the population, calculating their medians and generating a distribution of medians.



```python
def sample_median(population,sample_size, sample_count):
    """ Generate sampling distribution of medians 
    
    
    @param population is population distribution 
    @param sample_size is the number of samples 
    @param sample_count number of times samples are chosen

    
    """
    median_sample=[int(statistics.median(random.sample(population,sample_size))) for _ in range(sample_count)]
    sdmd_value,sdmd_freq=np.unique(median_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Medians, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(median_sample):.2f}, standard error={std_err:.2f}',sdmd_value,sdmd_freq)
    plt.show()
```

#### Graphs of different sample count
In this experiment, we first vary the sample size to study the variation in sampling distribution due to changes in sample count. We chose three different sample sizes: 30,10 and 100


```python
sample_median(uvs,30,10000)
```

```python
sample_median(uvs,10,10000)
```

```python
sample_median(uvs,100,10000)
```

#### Inference

From the above graphs, we can see that as the sample size increases, the sampling distribution closely resembles a normal distribution. The sampling distribution with sample size =100 is closest to a normal distribution and has mean closest to the population median.

#### Graphs of different sample counts
In the following experiment, we vary the sample count to see its effect on sampling distribution of medians. We chose sample sizes 100, 1000 and 20000 to study the effect of sample counts.

```python
sample_median(uvs,30,100)
```

```python
sample_median(uvs,30,1000)
```

```python
sample_median(uvs,30,20000)
```

#### Inference:
Clearly, from the above graphs we can infer that higher sample counts creates a sampling distribution closest to a normal distribution. However, a sample size of 30 and sample count 1000 has the sample mean closest to population median. 

### Sampling distribution of standard deviations
Here we generate a sample distribution of standard deviation by taking random samples from the population, calculating their standard deviation and generating a distribution of standard distributions.

```python
def sample_stddev(population,sample_size, sample_count):
    """ Generate sampling distribution of standard devaitions 
    
    
    @param population is population distribution 
    @param sample_size is the number of samples 
    @param sample_count number of times samples are chosen
    
    """

    
    stddev_sample=[int(statistics.pstdev(random.sample(population,sample_size))) for _ in range(sample_count)]
    stdev_value,stdev_freq=np.unique(stddev_sample,return_counts=True)
    std_err=statistics.pstdev(population)/math.sqrt(sample_size)
    create_bar_chart(f'Sampling distribution of Standard Deviations, sample count={sample_count},sample size={sample_size}, sample mean={statistics.mean(stddev_sample):.2f}, standard error={std_err:.2f}',stdev_value,stdev_freq)
    plt.show()
```

#### Graphs of different sample sizes
In the following experiment, we vary the sample size to see its effect on sampling distribution of std deviations. We chose sample sizes 10, 30 and 100 to study the effect of sample sizes.

```python
sample_stddev(uvs,10,10000)
```

```python
sample_stddev(uvs,30,10000)
```

```python
sample_stddev(uvs,100,10000)
```

#### Inference

From the above graphs, we see that sampling distributions of std deviations get centered around the population std deviation value as the sample size increase. The sample size of 100 sampling distribution appears closest to normal distriution and also has sample pean closest to population std deviation.

#### Graphs of different sample counts

Here, we vary the sample count to study the variation in sampling distribution of std deviations. We chose sample counts 100, 1000 and 10,000 and kept the sample size constant.

```python
sample_stddev(uvs,30,100)
```

```python
sample_stddev(uvs,30,1000)
```

```python
sample_stddev(uvs,30,20000)
```

#### Inference:

We can infer from the above graphs that as we increase sample count, the distribution approaches normal distribution. The sample cout 20,000 has closest resmblence to normal distribution.
