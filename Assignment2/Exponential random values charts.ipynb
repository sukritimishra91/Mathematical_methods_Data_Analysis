# Seaborn Charts of Exponentially Distributed Random Values

```python
import numpy as np
```

```python
import seaborn as sns
sns.set_context('poster')
sns.set_style('darkgrid')
```

```python
import random 
```

## Generate 1000 exponentially distributed random variables

```python
exponential_values=[int(random.expovariate(1/12.5)) for _ in range(1000)]
```

```python
values_exp,frequencies_exp=np.unique(exponential_values,return_counts=True)
```

## Seaborn Charts

### Bar Chart

```python
axes=sns.barplot(x=values_exp, y=frequencies_exp,color='c')
axes.set_title('Exponential Distribution')
axes.set(xlabel='Value', ylabel='Frequency')

sns.set(rc={'figure.figsize':(25,8)})
axes.set(xlim=(0,25))
```

### Univariate Distribution plot 

```python
axes=sns.distplot(exponential_values, color='c')
axes.set_title('Exponential Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
```

### Line plot

```python
axes=sns.lineplot(x=values_exp, y=frequencies_exp,color='c')
axes.set_title('Exponential Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
axes.set(xlim=(0,30))
```
