# Seaborn Charts of Uniform Random Values

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import random 
```

## Generating 1000 Uniform Random Values between 0 and 25

```python
sns.set(rc={'figure.figsize':(20,8)})
```

```python
uniform_values=[int(random.uniform(0,26)) for _ in range(1000)]
```

```python
print(np.mean(uniform_values).round(1))
```

```python
values,frequencies=np.unique(uniform_values,return_counts=True)
```

### Seaborn charts

#### Barchart

```python
axes=sns.barplot(x=values,y=frequencies,color='c')
axes.set_title('Uniform Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
sns.set_style('whitegrid')
sns.set_context('poster')
```

#### Line Graph

```python
axes=sns.lineplot(x=values,y=frequencies,color='c')
axes.set_title('Uniform Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
```

#### Univariate distribution plot (with histogram and kernal density estimates)

```python
axes=sns.distplot(values,color='c')
axes.set_title('Uniform Distribution')
axes.set(xlabel='Value', ylabel='Frequency')
sns.set_context('poster')
```
