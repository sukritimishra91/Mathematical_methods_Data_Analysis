## Load dataset to pandas dataframe ###

```python
import pandas as pd

car_data=pd.read_csv('Commercials_vehicles.csv')
```

```python
car_data.dtypes
```

### Filter all the cars that are discontinued in the market

```python
continued_car_data = car_data[car_data.Discontinued == False]
```

### Drop Unwanted Columns from the dataset

```python
continued_car_data=continued_car_data.drop(columns=['Discontinued','Boot_Space','Torque','Minimum_Turning_Radius','Steering_Type','Fuel_Tank_Capacity','Mileage_Highway','Mileage_City','Seating_Capacity','Rear_Brakes','Front_Brakes','Tyres','Wheel_Size','Wheel_Type','Link'])
```

### Sample view of the dataset

```python
continued_car_data.head()
```

### Data types of the columns

```python
continued_car_data.dtypes
```

## Cleaning dataset 

### Converting dataframe objects to string type

```python
continued_car_data=continued_car_data.astype(str)
```

### Splitting the Power column to seperate rpm

```python
car_data_new=continued_car_data["Power"].str.split("@",n=-1,expand = True)
```

```python
continued_car_data['Power']=car_data_new[0]
```

### Splitting the Manufacturer from model and version

```python
car_data_new=continued_car_data['Make_model_version'].str.split(' ',n=1,expand=True)
car_data_new
```

### Splitting data using string manipulation

```python
continued_car_data['Manufacturer']=car_data_new[0]
continued_car_data['Model_Version']=car_data_new[1]
continued_car_data.drop(columns=['Make_model_version','Ground_Clearance','Overall_Height'],inplace=True) 
```

```python
import numpy as np 
```

### Removing unwanted strings from data

```python
car_data_clean=continued_car_data.replace(['mm','Kg','cc','PS','kmpl','bhp',','],'', regex=True)
```

### Replacing empty values with NaN

```python
car_data_clean.replace('--',np.nan,inplace=True)
```

### Sample data set view after replacing empty values

```python
car_data_clean.head()
```

### Rename columns

```python
car_data_clean.rename(columns={'Price':'Price (INR)','Overall_Length':'Overall_Length (mm)','Overall_Width':'Overall_Width (mm)','Wheel_Base':'Wheel_base (mm)','Kerb_Weight':'Kerb_Weight (kg)','Displacement':'Displacement (cc)','Power':'Power (PS)','Mileage_Overall':'Mileage (kmpl)'},inplace=True) #renaming column headings to include units
```

```python
new_index=['Manufactuerer','Model_Version','Price (INR)','Overall_Length (mm)','Overall_Width (mm)','Wheel_base (mm)','Kerb_Weight (kg)','Displacement (cc)','Power (PS)','Valve_Mechanism','No_of_Cylinders','Cylinder_Configuration','Valves_per_Cylinder','Mileage (kmpl)','Transmission_Type']
car_data_clean.reindex(new_index,axis='columns')
```

### Converting column values to numeric values

```python
car_data_clean["Overall_Width (mm)"] = pd.to_numeric(car_data_clean["Overall_Width (mm)"],errors='coerce')
car_data_clean["Overall_Length (mm)"] = pd.to_numeric(car_data_clean["Overall_Length (mm)"],errors='coerce')
car_data_clean["Wheel_base (mm)"] = pd.to_numeric(car_data_clean["Wheel_base (mm)"],errors='coerce')
car_data_clean["Mileage (kmpl)"] = pd.to_numeric(car_data_clean["Mileage (kmpl)"],errors='coerce')
car_data_clean["Power (PS)"] = pd.to_numeric(car_data_clean["Power (PS)"],errors='coerce')
car_data_clean["Kerb_Weight (kg)"] = pd.to_numeric(car_data_clean["Kerb_Weight (kg)"],errors='coerce')
car_data_clean["Price (INR)"] = pd.to_numeric(car_data_clean["Price (INR)"],errors='coerce')
```

### Computing footprint of cars

```python
car_data_clean['Footprint (sqm)']=car_data_clean["Overall_Width (mm)"]*car_data_clean["Wheel_base (mm)"]
car_data_clean.loc[:,'Footprint (sqm)']*=0.86/1000000  #Calculating Footprint using mean track width multiplier
```

### Computing CO2 emissions from mileage using using conversion factor=23.7 g/l

```python
car_data_clean['CO2 Emission (g/km)']=car_data_clean['Mileage (kmpl)'].apply(lambda x: (100*x**-1)*23.71)
```

### Calculating the Power in kilowatts

```python
car_data_clean['Power (kW)']=car_data_clean.loc[:,'Power (PS)']*0.74
car_data_clean.drop(columns=['Power (PS)'],inplace=True)
```

### Setting precision of values in dataframe

```python
pd.set_option('precision',2)

car_data_clean.drop(car_data_clean.loc[car_data_clean['Manufacturer']=='nan'].index, inplace=True) #dropping unwanted rows
```

### Adding a column with car classification

```python
car_data_clean['Car Class']=(np.select([car_data_clean['Overall_Length (mm)'] < suv_length,(car_data_clean['Overall_Length (mm)'] >= suv_length) & (car_data_clean['Overall_Length (mm)'] < sedan_length),(car_data_clean['Overall_Length (mm)'] >= sedan_length) & (car_data_clean['Overall_Length (mm)'] < van_length)],['hatchback','suv','sedan'],default='van'))
```

#### Statistics of cars by class

```python
suv_car_class = car_data_clean[car_data_clean['Overall_Length (mm)'] == car_data_clean['Overall_Length (mm)']]
suv_car_class = suv_car_class[(suv_car_class['Overall_Length (mm)'] < sedan_length) & (suv_car_class['Overall_Length (mm)'] >= suv_length)]
suv_car_class.describe()
```

```python
sedan_car_class = car_data_clean[car_data_clean['Overall_Length (mm)'] == car_data_clean['Overall_Length (mm)']]
sedan_car_class = sedan_car_class[(sedan_car_class['Overall_Length (mm)'] >= sedan_length) & (sedan_car_class['Overall_Length (mm)'] < van_length)]
sedan_car_class.describe()
```

```python
van_car_class = car_data_clean[car_data_clean['Overall_Length (mm)'] == car_data_clean['Overall_Length (mm)']]
van_car_class = van_car_class[(van_car_class['Overall_Length (mm)'] >= van_length)]
van_car_class.describe()
```

```python
hatchback_car_class = car_data_clean[car_data_clean['Overall_Length (mm)'] == car_data_clean['Overall_Length (mm)']]
hatchback_car_class = hatchback_car_class[(hatchback_car_class['Overall_Length (mm)'] < suv_length)]
hatchback_car_class.describe()
```

### Cars distribution by class

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('poster')
plt.rcParams["figure.figsize"] = (20,10)

def show_bar_chart(title, x_labels, y_label, x_values, y_values, bar_topper):
    """
    Display a bar chart.
    @param title the chart title.
    @param x_labels the labels for the x axis
    @param y_label  the label for the y axis
    @param x_values the x values to plot
    @param y_values the y values to plot
    @param bar_text the text above each bar
    """
    
    sns.set_style('whitegrid')  # white backround with gray grid lines
    axes = sns.barplot(x_values, y_values, palette='bright')
    axes.set_title(title)
    axes.set(xlabel=x_labels, ylabel=y_label)

    # Scale the y-axis by 10% to make room for text above the bars.
    axes.set_ylim(top=1.10*max(y_values))

    # Display the percentage above each patch (bar).
    for bar, topper in zip(axes.patches, bar_topper):
        text_x = bar.get_x() + bar.get_width() / 2.0  
        text_y = bar.get_height() 
        axes.text(text_x, text_y, topper, 
                  fontsize=11, ha='center', va='bottom')

    plt.show()  # display the chart 
```

```python
suv_car_class_count = suv_car_class['Car Class'].count()
sedan_car_class_count = sedan_car_class['Car Class'].count()
van_car_class_count = van_car_class['Car Class'].count()
hatchback_car_class_count = hatchback_car_class['Car Class'].count()

total_car_class = suv_car_class_count + sedan_car_class_count + van_car_class_count + hatchback_car_class_count

pct_suv_car_class = 100*suv_car_class_count/total_car_class
pct_sedan_car_class = 100*sedan_car_class_count/total_car_class
pct_van_car_class  = 100*van_car_class_count/total_car_class
pct_hatchback_car_class  = 100*hatchback_car_class_count/total_car_class

classes  = ['Suv', 'Sedan', 'Van', 'Hatchback']
pcts     = [int(pct_suv_car_class),int(pct_sedan_car_class), int(pct_van_car_class), int(pct_hatchback_car_class)]
cars  = [f'{pcts[0]}%', f'{pcts[1]}%', f'{pcts[2]}%', f'{pcts[3]}%']

show_bar_chart('Percentage of car classes in total cars',
               'Car class', 'Percentage', 
               classes, pcts, cars)
```
